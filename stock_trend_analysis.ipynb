# 📈 Stock Market Trend Analysis - AAPL
This notebook analyzes AAPL stock trends using simulated data from 2020 to 2023.
- Metrics: Close price, Volume
- Technical Indicators: 20-day & 50-day Moving Averages


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('AAPL_stock_trend_analysis.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
# Plot closing price and moving averages
plt.figure(figsize=(14,6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA20'], label='20-Day MA')
plt.plot(df['MA50'], label='50-Day MA')
plt.title('AAPL Stock Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot trading volume
plt.figure(figsize=(14,4))
df['Volume'].plot(kind='bar', color='skyblue')
plt.title('AAPL Trading Volume (2020–2023)')
plt.ylabel('Volume')
plt.tight_layout()
plt.show()
```
